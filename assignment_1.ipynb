```python
import pandas
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
stu = pandas.read_csv("/content/StudentsPerformance.csv")
```

```python
plt.scatter(x=stu['math score'],y= stu['reading score'],color='red')
plt.xlabel("math score")
plt.ylabel("reading score")
plt.title("scatter plot between math score and reading score")
plt.show()
```

```python
plt.scatter(x=stu['math score'],y= stu['writing score'],color='blue')
plt.xlabel("math score")
plt.ylabel("writing score")
plt.title("scatter plot between math score and writting score")
plt.show()
```

```python
plt.scatter(x=stu['writing score'],y= stu['reading score'],color='pink')
plt.xlabel("writing score")
plt.ylabel("reading score")
plt.title("scatter plot between writing score and reading score")
plt.show()
```

```python
plt.boxplot(stu['math score'])
plt.xlabel("math score")
plt.title("math score")
plt.show()
```

```python
plt.boxplot(stu['reading score'])
plt.xlabel("reading score")
plt.title("reading score")
plt.show()
```

```python
plt.boxplot(stu['writing score'])
plt.xlabel("writing score")
plt.title('box plot for writing score')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='gender', y='math score', hue='race/ethnicity', data=stu)
plt.title("Math Scores by Gender and Race")
plt.show()

# Generate box plots for reading scores by gender and race
plt.figure(figsize=(12, 6))
sns.boxplot(x='gender', y='reading score', hue='race/ethnicity', data=stu)
plt.title("Reading Scores by Gender and Race")
plt.show()

# Generate box plots for writing scores by gender and race
plt.figure(figsize=(12, 6))
sns.boxplot(x='gender', y='writing score', hue='race/ethnicity', data=stu)
plt.title("Writing Scores by Gender and Race")
plt.show()
```
